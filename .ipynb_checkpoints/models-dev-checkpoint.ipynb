# DataLoader

```python
#include open price on the predicted day！！！！！！！！！
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
transformed['open_bid_pct'] = transformed['open_bid_pct'].shift(-1)
transformed['open_ask_pct'] = transformed['open_ask_pct'].shift(-1)
transformed = transformed.dropna()



train_data,valid_data,test_data=eda.dataset_split(transformed,0.7,0.2)


scaler = MinMaxScaler()

train_data_normalized = scaler.fit_transform(train_data[feat])
valid_data_normalized = scaler.transform(valid_data[feat])
test_data_normalized = scaler.transform(test_data[feat])


train_tensor = torch.Tensor(train_data_normalized)
valid_tensor = torch.Tensor(valid_data_normalized)
test_tensor = torch.Tensor(test_data_normalized)


window_size = 10


train_dataset = eda.MyDataset(train_tensor, window_size)
valid_dataset = eda.MyDataset(valid_tensor, window_size)
test_dataset = eda.MyDataset(test_tensor, window_size)


batch_size = 128
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
test_data.shape
```
